Diagnostics info on data processing

```python
%load_ext autoreload
%autoreload 2

import json
from pprint import pprint

import pandas as pd

from yiutils.project_utils import find_project_root
```

```python
proj_root = find_project_root("justfile")
print(proj_root)

data_dir = proj_root / "data" / "intermediate" / "llm-results-aggregated"
assert data_dir.exists()
```

----

# Raw data


## deekseek-r1-distilled


```python
data_path = data_dir / "deepseek-r1-distilled" / "raw_results.json"
assert data_path.exists()

df_ds_r1_raw = pd.read_json(data_path, orient="records")
df_ds_r1_raw.info()

df_ds_r1_raw
```

```python
# Basic Diagnostics
print("Number of rows:", len(df_ds_r1_raw), "\n")
print("Number of columns:", len(df_ds_r1_raw.columns), "\n")
print("Missing values per column:\n", df_ds_r1_raw.isnull().sum(), "\n")
print(
    "Number of unique pmid:",
    df_ds_r1_raw["pmid"].nunique()
    if "pmid" in df_ds_r1_raw.columns
    else "pmid not found",
    "\n",
)
if "pmid" in df_ds_r1_raw.columns:
    print(
        "Number of duplicated pmid:",
        df_ds_r1_raw.duplicated(subset=["pmid"]).sum(),
        "\n",
    )
print("Column types:\n", df_ds_r1_raw.dtypes, "\n")
```

### schema validation

```python
import jsonschema
```

```python
path_to_schema = (
    proj_root / "data" / "assets" / "data-schema" / "llama3" / "metadata.json.schema"
)
assert path_to_schema.exists(), f"Schema file not found at {path_to_schema}"
with open(path_to_schema, "r") as f:
    schema = json.load(f)
```

```python
data = df_llama3_raw["metadata"][0]
data
```

```python
jsonschema.validate(instance=data, schema=schema)
```
